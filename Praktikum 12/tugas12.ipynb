# **Import Library**

```python
#Import library untuk PCA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
```

# **Load Dataset**

```python
#Import dataset menggunakan link gdrive
df= pd.read_csv('https://drive.google.com/uc?export=download&id=1IZxa48F-yezCYnkU3qH4Rb7vdRdi6xHg')

#Menampilkan dataset
df.head()
```

```python
#Informasi dataset
df.info()
```

```python
#Statistik deskriptif
df.describe()
```

# **Data Preparation**

```python
#Cek missing value
df.isnull().sum()
```

```python
#Hapus kolom tidak digunakan
df = df.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
#Cek duplikat data
df.duplicated().sum()
```

```python
#Cek nilai unik kolom diagnosis
df['diagnosis'].value_counts()
```

```python
#Transformasi label diagnosis
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
#Memastikan label sudah di encode
df['diagnosis'].value_counts()
```

# **Seperating Dataset**

```python
#Pemisahan fitur dan target
x = df.drop(['diagnosis'], axis=1)
y = df['diagnosis']

#Menampilkan dimensi fitur dan target
print("Shape x:", x.shape)
print("Shape y:", y.shape)
```

# **Splitting Dataset**

```python
#Membagi data menjadi data train dan data test
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42, stratify=y)

print("Shape x_train:", x_train.shape)
print("Shape x_test:", x_test.shape)
```

# **Standarisasi Dataset**

```python
#Standarisasi dataset
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

x_train_scaled[:5]
```

# **Modelling**

## **SVM tanpa PCA**

```python
#Membuat model SVM
svm_nopca = SVC(kernel='rbf', gamma='scale', random_state=42)

#Melatih model SVM
svm_nopca.fit(x_train_scaled, y_train)

#Memprediksi dan evaluasi
y_pred_nopca = svm_nopca.predict(x_test_scaled)

#Akurasi
accuracy_nopca = accuracy_score(y_test, y_pred_nopca)
print("Akurasi No PCA:", accuracy_nopca)

#Laporan Klasifikasi
print("\nClasification Report No PCA")
print(classification_report(y_test, y_pred_nopca, target_names=['Benign', 'Malignant']))
```

## **SVM dengan PCA**

```python
#Membuat PCA dengan 3 komponen utama
PCA = PCA(n_components=3)

#Mengaplikasikan PCA pada dataset
x_train_pca = PCA.fit_transform(x_train_scaled)
x_test_pca = PCA.transform(x_test_scaled)

#Menampilkan dimensi PCA
print("Shape x_train_pca:", x_train_pca.shape)
print("Shape x_test_pca:", x_test_pca.shape)
```

```python
#Variasi setiap komponen PCA
explained_variance = PCA.explained_variance_ratio_

#Menampilkan rasio variasi
print("Rasio variasi komponen PCA:", explained_variance)
print("Total variasi             :", sum(explained_variance))
```

```python
#Visualisasi bar chart dari variasi 3 komponen utama
plt.bar([1,2,3], explained_variance)
plt.xlabel('Komponen Utama PCA')
plt.ylabel('Rasio Variasi')
plt.title('Rasio Variasi Setiap Komponen PCA')
plt.show()
```

```python
#Membuat model SVM dengan PCA
svm_pca = SVC(kernel='rbf', gamma='scale', random_state=42)

#Melatih model SVM dengan PCA
svm_pca.fit(x_train_pca, y_train)

#Memprediksi dan evaluasi
y_pred_pca = svm_pca.predict(x_test_pca)

#Akurasi
accuracy_pca = accuracy_score(y_test, y_pred_pca)
print("Akurasi PCA:", accuracy_pca)

#Laporan Klasifikasi
print("\nClasification Report PCA")
print(classification_report(y_test, y_pred_pca, target_names=['Benign', 'Malignant']))
```

# **Visualisasi 3D PCA**

```python
#Visualisasi 3D PCA 3 komponen utama
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(
    x_train_pca[:, 0],
    x_train_pca[:, 1],
    x_train_pca[:, 2],
    c=y_train,
    s=60
)

ax.set_title('Visualisasi PCA (3 Komponen) - Dataset Breast Cancer')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

#Tambah Legend berdasarkan Kelas
legend1 = ax.legend(*scatter.legend_elements(), title="Kelas")
ax.add_artist(legend1)

plt.show()
```

```python
#Perbandingan dengan PCA dan No PCA
comparison = pd.DataFrame({
    'Model': ["SVM No PCA", "SVM PCA (3 Komponen)"],
    'Jumlah Fitur': [x_train_scaled.shape[1], x_train_pca.shape[1]],
    'Akurasi': [accuracy_nopca, accuracy_pca],
    'Variansi Total PCA': [None, explained_variance.sum()]
})

comparison
```
